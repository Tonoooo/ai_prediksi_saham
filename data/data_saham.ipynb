Cara mendapatkan semua ticker perusahaan:

Pergi ke https://www.idx.co.id/data-pasar/data-saham/indeks-saham/   -> pilih kode indeks ISSI -> unduh file -> Buka file excel -> copi kolom indeks (dengan drag dari atas sampai bawah) -> paste di file txt.

~~ Perbedaan issi_...txt dengan ticker_issi.txt :

issi_....txt = ini berisi ticker issi yang belum disaring (isinya masih utuh). ini dipakai untuk mendapatkan data saham

ticker_issi.txt = ini berisi ticker yang sudah disaring, ticker-ticker ini berisi saham yang panjang data sahamnya lebih dari 200

!! PERINGATAN : Panjang data saham yang kurang dari 200 tidak akan disimpan data sahamnya

# Membaca setiap baris file dan Menyimpan setiap baris sebagai elemen di dalam list

```python
# menjadikan issi_...txt menjadi sebuah list
with open('issi_20220310.txt') as file:
    ticker = file.readlines()
    full_ticker = [line.rstrip() for line in ticker]
```

```python
print(len(full_ticker)) # banyak tickernya 469
full_ticker
```

# Membagi dua list ticker
karna batas output yfinance hanya sebanyak 281.
panjang full_ticker tadi harus dibagi dua list. list pertama berisi 1/2 ticker. list kedua berisi 1/2, tapi karna ticker TRIL tidak ada diyfinance jadi di list ticker kedua hapus TRIL

```python
# mengambil tickernya dari awal sampai sebelum 239 jadi dari indeks 0 sampai 238. 
ticker1 = full_ticker[:int(len(full_ticker)/2)+1] 
# jumlah banyak ticker nya jadi 239
print(len(ticker1))
```

```python
# dari 239 sampai akhir
ticker2 = full_ticker[int(len(full_ticker)/2)+1:] 
print(len(ticker2))
```

```python
## Menghapus ticker TRIL di list ticker2
#karna sepertinya TRIL tidak ada di yahoo finance
ticker2.remove('TRIL') 
len(ticker2)
```

# Mendapatkan dan Menyimpan data saham

```python
import pandas_datareader as pdr
import pandas as pd

# tentukan awal dan akhir data nya
mulai = '1980-01-01'
akhir = '2022-03-11'

#Tempat bagi saham yang berukuran kecil, dan saham ini tidak akan disimpan data nya
saham_kecil = []
```

```python
#---------------- for loop untuk mendapatkan data data nya dan disimpan difiles csv
## bagian pertama
for tic in ticker1:
    # dapatkan data nya
    df = pdr.DataReader(f'{tic}.JK', data_source='yahoo',start=f'{mulai}', end=f'{akhir}') # dapatkan datanya
    if len(df) >= 200: # jika panjang data sahamnya lebih dari 200 maka :
        df.to_csv(f'bsaham/{tic}.csv')# simpan data nya
    else: # jika panjang data sahamnya kurang dari 200 maka :
        saham_kecil.append(tic)

## bagian kedua
for tic in ticker2:
    # dapatkan data nya
    df = pdr.DataReader(f'{tic}.JK', data_source='yahoo',start=f'{mulai}', end=f'{akhir}') # dapatkan datanya
    if len(df) >= 200:
        df.to_csv(f'bsaham/{tic}.csv')# simpan data nya
    else:
        saham_kecil.append(tic)
```

# Menghapus ticker yg samadengan saham_kecil di ticker_issi.txt

setelah kita mendapatkan list ticker dengan panjang data sahamnya kurang dari 200 (saham_kecil), sekarang kita hapus di ticker_issi.txt yang mempunyai ticker yang sama dengan saham_kecil.

ini berguna saat melatih/fit ai kita bersumber dari ticker, ia tidak akan mendapatkan saham dengan panjang datanya kurang. jadi ai nya tidak kekurangan data latih

jadi ticker_issi.txt inilah yang akan menjadi sumber ticker untuk ai berlatih

```python
len(saham_kecil)
```

```python
# jadikan ticker_issi.txt menjadi sebuah list
with open('ticker_issi.txt') as file:
    tickernya = file.readlines()
    list_ticker_issi = [line.rstrip() for line in tickernya]

# hapus saham TRIL di list_ticker_issi (Karna TRIL tidak ada di yahoofinance)
list_ticker_issi.remove('TRIL')

# hapus ticker saham di list_ticker_issi jika ada yang ticker sahamnya samadengan yang ada 
#di saham_kecil
for i in saham_kecil:
    list_ticker_issi.remove(i)

# sekarang jadikan kembali ticker_issi itu menjadi txt seperti semula
ticker_issi_file = open("ticker_issi.txt", "w")
for element in list_ticker_issi:
    ticker_issi_file.write(element + "\n")
ticker_issi_file.close()
```

```python
# menyimpan saham_kecil menjadi file txt (untuk dokumentasi saja)
saham_kecil_file = open("saham_kecil.txt", "w")
for element in saham_kecil:
    saham_kecil_file.write(element + "\n")
saham_kecil_file.close()
```
