```python
import pandas as pd
from fungsi import datanya
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization, LSTM
from keras.callbacks import ModelCheckpoint
```

```python
window_size = 20 # timestep

# ~~~~~ file yang berisi ticker jadikan sebuah list
with open('../data/ticker_issi.txt') as file:
    tick = file.readlines()
    ticker = [line.rstrip() for line in tick]# ini adalah list yang berisi ticker
```

```python
# hari pertama
ticker = ticker[:149]

# hari kedua
# ticker = ticker[149:298]

# hari ketiga
# ticker = ticker[298:]
```

```python
for tic in ticker:
    data = pd.read_csv(f'../data/bsaham/{tic}.csv')
    X_train, Y_train = datanya.data_latih4(data, window_size)
    model = Sequential()
    model.add(LSTM(units = 80, activation = 'relu', return_sequences = True, 
                input_shape = (X_train.shape[1], window_size)))
    model.add(Dropout(0.2))
    model.add(BatchNormalization())
    model.add(LSTM(units = 44,return_sequences=True))
    model.add(Dropout(0.2))
    model.add(BatchNormalization())
    model.add(LSTM(units = 44,return_sequences=True))
    model.add(Dropout(0.2))
    model.add(BatchNormalization())
    model.add(LSTM(units = 44,return_sequences=True))
    model.add(Dropout(0.2))
    model.add(BatchNormalization())
    model.add(LSTM(units = 17,activation='relu'))
    model.add(Dropout(0.2))
    model.add(BatchNormalization())
    model.add(Dense(1, activation = 'linear'))
    model.compile(loss = 'mean_squared_error', optimizer = 'adam')
    filepath = f"model/{tic}_model.h5"
    checkpoint = ModelCheckpoint(filepath, monitor='loss', verbose=0, save_best_only=True, mode='min')
    callbacks_list = [checkpoint]
    model.fit(X_train, Y_train, epochs=50, batch_size=20, shuffle=False, callbacks=callbacks_list)
```
